```python
import numpy as np
import math as mt
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv('./Salary_Data.csv')
x_train=df['YearsExperience'].to_numpy()
y_train=df['Salary'].to_numpy()
plt.plot(x_train,y_train)
plt.xlabel("Years Experience")
plt.ylabel("Salary")
plt.title("Experience vs Salary")
plt.show()
```

```python
def gradient(x,y,w,b):
    m=x.shape[0]
    dersum_w=np.float64(0.0)
    dersum_b=np.float64(0.0)
    for i in range (m):
        f_wb=np.float64(w*x[i]+b)
        err=np.float64(f_wb-y[i])
        der_w=np.float64(err*x[i])
        der_b=np.float64(err)
        dersum_w+=der_w
        dersum_b+=der_b
    return dersum_w,dersum_b 
```

```python
def calculate_descent(x_train,y_train,w_in,b_in,alpha,iters,m):
    w=w_in
    b=b_in
    for i in range(iters):
        deri_w,deri_b=gradient(x_train,y_train,w,b)
        temp_w=w-alpha*deri_w
        temp_b=b-alpha*deri_b
        w=temp_w
        b=temp_b
        if i%1000==0:
            y_temp=[]
            for i in range(m):
                y_val=w*(x_train[i])+b
                y_temp.append(y_val)
            plt.subplot(1,3,1)    
            plt.plot(x_train,y_temp)
            plt.title("Gradient descent")                   
    return w,b  
```

```python
w_in=0
b_in=0
m=x_train.shape[0]
iters=10000
alpha=0.001
fin_w,fin_b=calculate_descent(x_train,y_train,w_in,b_in,alpha,iters,m)
y_tempi=[]
for i in range(m):
    y_val=fin_w*(x_train[i])+fin_b
    y_tempi.append(y_val)
plt.subplot(1,3,2)
plt.plot(x_train,y_tempi,'xr')
plt.title("Model Predictions")  
plt.subplot(1,3,3)
plt.plot(x_train,y_train)
plt.title("Training data")
print(f"The salary of employee with 1.1 year experience is: $ {fin_w*1.1+fin_b:0.1f}")
print(f"The salary of employee with 1.5 year experience is: $ {fin_w*1.5+fin_b:0.1f}")
print(f"The salary of employee with 3 year experience is: $ {fin_w*3.0+fin_b:0.1f}")
print(f"The salary of employee with 7 year experience is: $ {fin_w*7.0+fin_b:0.1f}")
print(f"The salary of employee with 6 year experience is: $ {fin_w*6.0+fin_b:0.1f}")
```
